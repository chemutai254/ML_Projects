```python
# Feature 'left' split over the feature 'salary'
sns.catplot(x='left', col='salary', kind='count', data=df1);
```

```python
# Feature 'left' split over the feature 'promotion_last_5years'
sns.catplot(x='left', col='promotion_last_5years', kind='count', data=df1);
```
